```python
import os
import warnings

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm

from utils.config import data_path

warnings.filterwarnings("ignore")
%matplotlib inline
plt.style.use('seaborn-white')
```

```python
# pipeline
def simple_ols(df, x , y):
    reg = smf.ols(formula=f'{y} ~ {x}', data=df).fit()
    print(reg.summary())
    print(anova_lm(reg))
    print(f'\n Correlation Coef: \n {np.corrcoef(df[x], df[y])[0, 1]}')
```

## 2.1 Inference for the population Intercept and slope

```python
# skin cancer data
df = pd.read_csv(os.path.join(data_path, 'skincancer.txt'), delimiter='\t')
df.head(5)

# plot
sns.regplot(x='Lat', y='Mort', data=df, line_kws={'color': 'red'})

# regression
simple_ols(df, x='Lat', y='Mort')
```

## 2.2 Another example of slop interference

```python
# lead cord data
df = pd.read_csv(os.path.join(data_path, 'leadcord.txt'), delimiter='\t')
df.head()

# plot
sns.regplot(x='Sold', y='Cord', data=df)

# reg
simple_ols(df, x='Sold', y='Cord')
```
